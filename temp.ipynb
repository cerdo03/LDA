```python
import nltk
```

```python
f1=open("plot_summaries.txt",encoding="utf8")
s=f1.readlines()
len(s)
```

```python
st=s[0:5]
len(st)
```

```python
st
```

```python
#Stop Words

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
 
newlist=[]
for si in st:
    stop_words = set(stopwords.words('english'))
    
    word_tokens = word_tokenize(si)
    
    tokens = [token.lower() for token in word_tokens]
    
    for w in tokens:
        if w not in stop_words:
            newlist.append(w)
 
#print(word_tokens)
print(newlist)
```

```python
#Porter Stemmer Steps and noun removal
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
  
l=[]
ps = PorterStemmer()

for si in newlist:
    tagged_sentence = nltk.tag.pos_tag(si.split())
    edited_sentence = [word for word,tag in tagged_sentence if tag != 'NNP' and tag != 'NNPS' and tag!='NN' and tag!='NNS']
    str=' '.join(edited_sentence)

    words=word_tokenize(str)
    for w in words:
        l.append(ps.stem(w))
len(l)
l
```

```python
f4=open("3.txt","a+")
for li in l:
    f4.write(li+" ")
```
