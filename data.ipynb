```python
import nltk
```

```python
f1=open("plot_summaries.txt",encoding="utf8")
s=f1.readlines()
len(s)
```

```python
#Stop Words

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
 
newlist=[]
for si in s:
    stop_words = set(stopwords.words('english'))
    
    word_tokens = word_tokenize(si)
    
    tokens = [token.lower() for token in word_tokens]
    tempList=[]
    for w in tokens:
        if w not in stop_words:
            tempList.append(w)
    newlist.append(tempList)
#print(word_tokens)
#print(newlist)
newlist[1]
```

```python
#Porter Stemmer Steps and noun removal
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
  
ps = PorterStemmer()

for i in range(0,len(newlist)):
    tempList=[]
    for si in newlist[i]:
        tagged_sentence = nltk.tag.pos_tag(si.split())
        edited_sentence = [word for word,tag in tagged_sentence if tag != 'NNP' and tag != 'NNPS' and tag!='NN' and tag!='NNS']
        s=' '.join(edited_sentence)

        words=word_tokenize(s)
        for w in words:
            tempList.append(ps.stem(w))
    #removing punctuations
    tempList = [word for word in tempList if word.isalpha()]

    newlist[i]=tempList
newlist
#l
```

```python
arr= []
for i in range(0,len(newlist)):
  for j in newlist[i]:
    arr.append(j)
arr
```

```python
import numpy as np
import pandas as pd



df=pd.DataFrame(arr)
df
```

```python
df[0].value_counts()
```

```python
df=df.groupby([0]).filter(lambda x: len(x) > 0).drop_duplicates(subset=0)
df
```

```python
hset=set()
for i in df[0]:
    hset.add(i)
hset
```

```python
dff=pd.DataFrame(hset)
dff['ind'] = range(1, len(df) + 1)
dff
```

```python
dic={}
for i in range(0,len(dff)):
  dic[dff[0][i]]=dff["ind"][i]
dic
```

```python
fl=[]
for i in range(0,len(newlist)):
    temp=[]
    for j in newlist[i]:
        if j in hset:
            temp.append(dic[j])
    fl.append(temp)
fl
```

```python
len(fl)
```

```python
final=[]
for i in range(0,len(fl)):
  tempDic={}
  for j in fl[i]:
    if j in tempDic.keys():
      tempDic[j]+=1
    else:
      tempDic[j]=1
  final.append(tempDic)
  
```

```python
final
```

```python
file = open("lda_data.dat","w")
```

```python
strings_list=[]
for doc in range(0,len(final)):
  str_to_add =  str(len(final[doc]))
  for word in final[doc]:
    str_to_add=str_to_add+" " + str(word) + ":"+ str(final[doc][word])
  str_to_add=str_to_add+" \n"
  strings_list.append(str_to_add)
```

```python
strings_list[0:3]
```

```python
file.writelines(strings_list)  
file.close()
```
